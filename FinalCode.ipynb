MLB attendance is missing Home_Percentage, Road_Percentage, and Total_Percentage values for the 2019 season. We can calculate what Home_Percentage should have been if we know venue capacities since all home games are played in 1 venue. To impute venue capacity data which we don't have, we'll find the 2018 venue capacities by taking the average attendance at home games and dividing by Home_Percentage variable and use these to calculate 2019 Home_Attendance. This works off the assumption that venue capacities did not change between 2018 and 2019. Although not a perfect methodology, there were no major venue changes between the 2018 and 2019 seasons in Major League Baseball, so any changes that did occurr to venue capacity should be minimal.

```python
# Calculate a Home_Capacity variable based on the total home attendance
MLB_attendance18 = MLB_attendance[MLB_attendance['Year']==2018]
MLB_attendance18['2018_Capacity'] = round(MLB_attendance18['Home_Average'] / (.01*MLB_attendance18['Home_Percentage']))
MLB_attendance18 = MLB_attendance18[['Team','2018_Capacity']]
MLB_attendance2 = MLB_attendance.merge(MLB_attendance18, on='Team', how='left')

MLB_attendance2.head(30) # let's check that capacity was added correctly
```

```python
MLB_attendance2['Orig_Home_Percentage'] = MLB_attendance2['Home_Percentage'] # to verify we fixed missing values correctly
MLB_attendance2['Home_Percentage'] = np.where(MLB_attendance2['Year']==2019,
                                                    round(MLB_attendance2['Home_Attendance'] / MLB_attendance2['2018_Capacity'],1),
                                                    MLB_attendance2['Home_Percentage'])
MLB_attendance2.head(50)
```

**Data dictionary:**
- `League`: Categorical: The professional sports league that each team belongs to. Values are MLB, NFL, NBA, NHL
- `Year`: Int: Years go from 2010-2019
- `Team`: Categorical: The team. 30 possibilities for MLB and NBA, and 32 for NFL and NHL
- `Home_Games`: Int: The number of home games throughout the year. Not more than 82 for MLB, 50 for NBA and NHL and 10 for NFL. Low values indicate teams have just moved to new stadium 
- `Home_Attendance`: Int: Total number of people that came to home games for that given year. Depending on the league, can be in the millions. Not more than 4,000,000 in our dataset
- `Home_Average`: Int: Average amount of people at each home game. Below 20,000 for NBA and NHL. Up to 50,000 or 60,000 for MLB, NFL
- `Home_Percentage`: Float: Home_Average divided by capacity of the stadium. Can be more than 100 if a team oversold their tickets. Generally not less than 20
- `Team_Year`: Categorical: Combines Year and Team column
- `W`: Int: The number of wins in that season. Not more than 108 for MLB, 16 for NFL, 70 for NBA and NHL
- `L`: Int: The number of losses in that season. Generally cannot be greater than 115 for MLB, 16 for NFL, 70 for NBA and NHL
- `T_or_OTL`: Float: The number of ties in that season. Between 0 and 18. Always 0 for NBA and MLB, generally not more than 3 for NFL
- `Market`: Categorical: A variable that accounts for the city or area the team is in. Accounts for the fact that some cities have more than one team and maybe even more than one team in a specific league

```python
# Now sort the values for ones where it is impractical to find unique values. 

data.sort_values("Home_Percentage")
data.sort_values("Home_Attendance")
data.sort_values("Home_Average")
data.sort_values("W")
data.sort_values("L")
data.sort_values("T_or_OTL")
```

```python
# There's something weird about some of the percentages in Home_Percentage. Below 1% doesn't make sense. Some research showed that those values
# were actually decimals and need to be changed. A team never has below 5% in our data, so anything that comes up as such
# must be a mistake. We change that
data['Home_Percentage'] = data['Home_Percentage'].apply(lambda x: x*100 if x <= 5 else x)

# Check to make sure it worked
data.sort_values("Home_Percentage")
```

## Data Vizualization and Study of Correlations between Attributes

```python
# Assigning 'teams.csv' data to dataset varibale 'df'
df = data 
df.head()
```

```python
df=df.drop(['Unnamed: 0'], axis=1)
```

```python
# Plotting Histograms for each variable 
sns.histplot(data=df, x='Home_Attendance')
```

Skewed Right

```python
sns.histplot(data=df, x='Home_Average')
```

Skewed Right

```python
plt.hist(df['Home_Percentage'])
plt.show()
```

```python
# Create Win % column
df['win_pct'] = df['W']/(df['W'] + df['L'])
```

```python
df.loc[df['League'] == 'NHL', 'win_pct'] = ((2*df['W'])+df['T_or_OTL'])/(2*(df['W']+df['L']+df['T_or_OTL']))
```

```python
sns.histplot(data=df, x='win_pct')
```

```python
# Standardizing win %
league_wpct = df.groupby(['League'])['win_pct'].describe()
```

```python
league_wpct
```

```python
wp_mean_mlb = league_wpct.iloc[0,1]
wp_sd_mlb = league_wpct.iloc[0,2]
wp_mean_nba = league_wpct.iloc[1,1]
wp_sd_nba = league_wpct.iloc[1,2]
wp_mean_nfl = league_wpct.iloc[2,1]
wp_sd_nfl = league_wpct.iloc[2,2]
wp_mean_nhl = league_wpct.iloc[3,1]
wp_sd_nhl = league_wpct.iloc[3,2]
```

```python
df.loc[df['League'] == 'MLB', 'win_pct'] = (df['win_pct'] - wp_mean_mlb)/wp_sd_mlb
df.loc[df['League'] == 'NBA', 'win_pct'] = (df['win_pct'] - wp_mean_nba)/wp_sd_nba
df.loc[df['League'] == 'NFL', 'win_pct'] = (df['win_pct'] - wp_mean_nfl)/wp_sd_nfl
df.loc[df['League'] == 'NHL', 'win_pct'] = (df['win_pct'] - wp_mean_nhl)/wp_sd_nhl
df
```

```python
# Standardizng Home_Percentage
league_hpct = df.groupby(['League'])['Home_Percentage'].describe()
league_hpct
```

```python
hp_mean_mlb = league_hpct.iloc[0,1]
hp_sd_mlb = league_hpct.iloc[0,2]
hp_mean_nba = league_hpct.iloc[1,1]
hp_sd_nba = league_hpct.iloc[1,2]
hp_mean_nfl = league_hpct.iloc[2,1]
hp_sd_nfl = league_hpct.iloc[2,2]
hp_mean_nhl = league_hpct.iloc[3,1]
hp_sd_nhl = league_hpct.iloc[3,2]
```

```python
df['hpct_std'] = df['Home_Percentage']
```

```python
df.loc[df['League'] == 'MLB', 'hpct_std'] = (df['hpct_std'] - hp_mean_mlb)/hp_sd_mlb
df.loc[df['League'] == 'NBA', 'hpct_std'] = (df['hpct_std'] - hp_mean_nba)/hp_sd_nba
df.loc[df['League'] == 'NFL', 'hpct_std'] = (df['hpct_std'] - hp_mean_nfl)/hp_sd_nfl
df.loc[df['League'] == 'NHL', 'hpct_std'] = (df['hpct_std'] - hp_mean_nhl)/hp_sd_nhl
```

```python
df.groupby(['League'])['hpct_std'].describe()
```

```python
# Plotting Attendence and Winning 
plt.scatter(df['Home_Percentage'], df['win_pct'], c=df['League'].astype('category').cat.codes)
leagues = ['MLB', 'NBA', 'NFL', 'NHL']
plt.xlabel("Home Percentage")
plt.ylabel("Win Percentage")
plt.legend(handles=plt.scatter(df['Home_Percentage'], df['win_pct'], c=df['League'].astype('category').cat.codes).legend_elements()[0], 
           labels = leagues,
           title="League")
plt.show()
```

```python
plt.scatter(df['hpct_std'], df['win_pct'], c=df['League'].astype('category').cat.codes)
leagues = ['MLB', 'NBA', 'NFL', 'NHL']
plt.xlabel("Home Percentage (Standardized)")
plt.ylabel("Win Percentage")
plt.legend(handles=plt.scatter(df['hpct_std'], df['win_pct'], c=df['League'].astype('category').cat.codes).legend_elements()[0], 
           labels = leagues,
           title="League")
plt.show()
```

```python
df_mlb = df[df['League']=='MLB']
df_nba = df[df['League']=='NBA']
df_nfl = df[df['League']=='NFL']
df_nhl = df[df['League']=='NHL']
```
